# Exploratory Data Analysis

```python
## 1 The statistical summary of earthquakes over the years
# Maximum magnitude (max) → the strongest earthquake in that year.
# Minimum magnitude (min) → the weakest earthquake recorded in that year.
# Mean magnitude (avg) → the average strength of all earthquakes in that year.

# Ensure magnitudes are numeric
mysql_df['magnitude'] = pd.to_numeric(mysql_df['magnitude'], errors='coerce')

# Filter unrealistic values (keep earthquakes 0–12 only)
mysql_df = mysql_df[(mysql_df['magnitude'] >= 0) & (mysql_df['magnitude'] <= 12)]

# Set global display format (2 decimal places with commas)
pd.options.display.float_format = '{:,.2f}'.format

# Extract year
mysql_df['occurred_on'] = pd.to_datetime(mysql_df['occurred_on'], errors='coerce')
mysql_df['year'] = mysql_df['occurred_on'].dt.year

# Group yearly stats
yearly_range = mysql_df.groupby('year')['magnitude'].agg(['min', 'max', 'mean'])
print(yearly_range.tail(20))
```

```python
#2 --- Insights from the yearly trend(Earthquakes per year(1969-2018) )---

# Count earthquakes per year
yearly_counts = mysql_df['year'].value_counts().sort_index()

# Now your analysis works
print("\n--- Insights on Earthquakes per Year (1969–2018) ---\n")

# Total earthquakes
print(f"Total earthquakes recorded: {yearly_counts.sum()}")

# Peak year
peak_year = yearly_counts.idxmax()
print(f"Peak year: {peak_year} with {yearly_counts.max()} earthquakes")

# Lowest year
low_year = yearly_counts.idxmin()
print(f"Lowest year: {low_year} with {yearly_counts.min()} earthquakes")

# Average per year
print(f"Average earthquakes per year: {round(yearly_counts.mean(),2)}")

# Trend (last year vs first year)
growth = ((yearly_counts.iloc[-1] - yearly_counts.iloc[0]) / yearly_counts.iloc[0]) * 100 if yearly_counts.iloc[0] != 0 else None
if growth is not None:
    print(f"Change from {yearly_counts.index[0]} to {yearly_counts.index[-1]}: {growth:.2f}%")
else:
    print("Trend change could not be calculated (first year has zero records).")
```

```python
#3 --- Insights on Earthquake Causes---
print("\n--- Insights on Earthquake Causes ---")

# Count earthquakes by cause
cause_counts = mysql_df['cause'].value_counts()

# Total earthquakes classified by cause
print(f"Total earthquakes classified by cause: {cause_counts.sum()}")

# Top cause
top_cause = cause_counts.idxmax()
print(f"Most common cause: {top_cause} ({cause_counts.max()} cases, "
      f"{(cause_counts.max()/cause_counts.sum()*100):.1f}% of total)")

# Least cause
least_cause = cause_counts.idxmin()
print(f"Least common cause: {least_cause} ({cause_counts.min()} cases, "
      f"{(cause_counts.min()/cause_counts.sum()*100):.1f}% of total)")

# Distribution check
print("\nCause distribution (percentage):")
print((cause_counts / cause_counts.sum() * 100).round(2))
```

```python
#4 Insights: extremes(Strongest and Deepest )

# Extract strongest and deepest earthquakes
strongest_eq = mysql_df.loc[mysql_df['magnitude'].idxmax(),['occurred_on','place','magnitude','depth']]
deepest_eq = mysql_df.loc[mysql_df['depth'].idxmax(),['occurred_on','place','magnitude','depth']]
print("\nStrongest Earthquake Overall:")
# print(mysql_df.loc[mysql_df['magnitude'].idxmax(), ['occurred_on','place','magnitude','depth']])
print(strongest_eq)

print("\nDeepest Earthquake Overall:")
# print(mysql_df.loc[mysql_df['depth'].idxmax(), ['occurred_on','place','magnitude','depth']])
print(deepest_eq)
```

```python
#5 Insight: top locations with most earthquakes
print("\nTop 5 Places with Most Earthquakes:")
print(mysql_df['place'].value_counts().head(5))
```

```python
#1  Yearly stats on the weakest,strongest,average earthquakes

plt.figure(figsize=(12,6))

# Plot lines
plt.plot(yearly_range.index, yearly_range['min'], label='Weakest (Min)', color='blue', linestyle='--', marker='o')
plt.plot(yearly_range.index, yearly_range['max'], label='Strongest (Max)', color='red', linestyle='--', marker='o')
plt.plot(yearly_range.index, yearly_range['mean'], label='Average (Mean)', color='orange', linewidth=2, marker='s')

plt.title("Yearly Earthquake Magnitude Trends (Weakest–Strongest–Average)")
plt.xlabel("Year")
plt.ylabel("Magnitude")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("Yearly_stats_on_earthquakes.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#2 Earthquakes per year(1969-2018)

plt.figure(figsize=(12,6))
mysql_df['year'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title("Earthquakes per Year (1969–2018)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("Earthquakes_per_year.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#3 Earthquake Causes

plt.figure(figsize=(10,6))
cause_counts = mysql_df['cause'].value_counts()

# Plot
cause_counts.plot(kind='bar', color='royalblue')
plt.title("Earthquake Causes")
plt.xlabel("Cause")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.savefig("Earthquake_Causes.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#4 Insights: extremes(Strongest and Deepest )
# Prepare data
labels = ['Strongest EQ', 'Deepest EQ']
magnitudes = [strongest_eq['magnitude'], deepest_eq['magnitude']]
depths = [strongest_eq['depth'], deepest_eq['depth']]
places = [strongest_eq['place'], deepest_eq['place']]

# Plot
fig, ax = plt.subplots(figsize=(8,6))

bar1 = ax.bar(labels, magnitudes, color='royalblue', width=0.4, label="Magnitude")
ax.set_ylabel("Magnitude", color='tomato')
ax.set_ylim(0, max(magnitudes) + 1)

# Annotate places
for i, place in enumerate(places):
    ax.text(i, magnitudes[i] + 0.1, f"{place}", ha='center', fontsize=9, wrap=True)

ax.set_title("Strongest and Deepest Earthquakes with Locations")
plt.savefig("Strongest_and_Deepest.png", dpi=300, bbox_inches="tight")
plt.show()

```

```python
#5 Insight: top5 locations with most earthquakes

top_places = mysql_df['place'].value_counts().head(5)

# Plot horizontal bar chart
plt.figure(figsize=(8,6))
top_places.plot(kind='barh', color='teal')

plt.xlabel("Number of Earthquakes")
plt.ylabel("Place")
plt.title("Top 5 Places with Most Earthquakes")
plt.gca().invert_yaxis()  # So the highest value appears on top

# Annotate counts on bars
for index, value in enumerate(top_places):
    plt.text(value + 0.5, index, str(value), va='center')
plt.savefig("top5_locations_with_most_earthquakes.png", dpi=300, bbox_inches="tight")
plt.show()
```
