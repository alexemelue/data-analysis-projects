```python
# importing the necessary modules for operation
import psycopg2 as pgconnect
import pandas as pd
import matplotlib.pyplot as plt 
from sqlalchemy import create_engine

import warnings 
warnings.filterwarnings("ignore")
```

```python
# get user password for database entry
import getpass
# Prompt the user for a password
passwd = getpass.getpass('Enter your PostgreSQL database password: ')
```

```python
#2 connecting to the PostgreSQL using sqlalchemy
import time

start_time = time.time()
table_name = 'airline_data_tb'
hostname = "localhost"
dbname = "Kc_Airline_Dataset_db"
username = "postgres"
port = 5432
engine = create_engine(f"postgresql+psycopg2://{username}:{passwd}@{hostname}:{port}/{dbname}")
print('Postgres Database is connected successfully!')
```

```python
# Write a SQL query to extract the required data
query = ''' 
SELECT *
FROM airline_data_tb;
'''
```

```python
# read and store the required table as pandas dataframe
postgresql_df = pd.read_sql(query,engine)
postgresql_df.head()
stop_time = time.time()
print(f'The query took {round(stop_time-start_time)}secs to extract the datas from the database to the dataframe')
```

```python
"""You are to provide the management of the airline data analysis of the dataset 
and also provide some visual representation, 
 write at least 6 python code blocks/statements to extract the required information """
```

```python
 #1 Total Count of passengers by gender

gender_counts = postgresql_df["Gender"].value_counts().reset_index()
print("Passenger count by Gender:")
gender_counts.head()
```

```python
import math
#2 Average passenger age
# pd.options.display.float_format = '{:,.2f}'.format # Show numbers in 2 d.p
avg_age = postgresql_df["Age"].mean()
# print("Average passenger age:",avg_age)
print("Average passenger age:", math.floor(avg_age))
```

```python
#3 Number of flights per continent
flights_by_continent = postgresql_df["Airport_Continent"].value_counts().reset_index()
print("Flights per Continent:")
flights_by_continent.head(10)
```

```python
#4 Nationality distribution of passengers

nationality_counts = postgresql_df["Nationality"].value_counts().head(10).to_frame("count")
print("\nTop 10 Passenger Nationalities:")
nationality_counts.head(10)
```

```python
#5 Flight status distribution (e.g., On Time, Delayed, Cancelled)
status_counts = postgresql_df["Flight_Status"].value_counts().to_frame("count")
print("\nFlight Status distribution:")
status_counts.head()
```

```python
#6 Most frequently used airports

airport_usage = postgresql_df["Airport_Name"].value_counts().head(10)
print("\nTop 10 Airports by Usage:" )
airport_usage.head(10)
```

```python
#7.cancelled flights based on season

postgresql_df["Date"] = pd.to_datetime(postgresql_df["Departure_Date"])

# Map months to seasons
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

postgresql_df["Season"] = postgresql_df["Date"].dt.month.apply(get_season)

# Count cancelled flights per season
cancelled_season = postgresql_df[postgresql_df["Flight_Status"] == "Cancelled"].groupby("Season").size().sort_values()

# converting the season data type to string
postgresql_df["Season"] =  postgresql_df["Season"].astype('string')

# calling the dataframe to display result
cancelled_season.to_frame("count")
```

```python
#8 top 10 business flight routes
routes = postgresql_df.groupby(["Airport_Name", "Arrival_Airport"]).size().reset_index(name="Flights")
top_routes = routes.sort_values("Flights", ascending=False).head(10)

print(top_routes)
```

```python
#9 flight distribution by seasons
season_counts = postgresql_df["Season"].value_counts().sort_values(ascending=False).to_frame('count')
season_counts
```

```python
# *********************Visualization Chats****************************************
```

```python
 #1 gender distribution

plt.figure(figsize=(6,4))
plt.bar(gender_counts["Gender"], gender_counts["count"], color=["steelblue", "lightcoral"])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
# plt.savefig('gender distribution.tiff')
plt.savefig("gender_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
##2 passenger age distribution

# Create bins for age
bins = [0, 18, 30, 45, 60, 100]
labels = ["<18","18-30","31-45","46-60","60+"]
postgresql_df["Age_Group"] = pd.cut(postgresql_df["Age"], bins=bins, labels=labels, right=False)

age_group_counts = postgresql_df["Age_Group"].value_counts().sort_index()

plt.figure(figsize=(8,5))
age_group_counts.plot(kind="bar", color="royalblue", edgecolor="black")
plt.title("Passenger Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Passengers")
# plt.savefig('passenger age distribution.tiff')
plt.savefig("passenger_age_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
##3 Number of flights per continent


flights_by_continent.set_index("Airport_Continent")["count"].plot(
    kind="pie", 
    autopct="%1.1f%%", 
    figsize=(6,6), 
    title="Flights per Continent"
)

plt.ylabel("")  # removes default ylabel
# plt.savefig('Number of flights per continent.tiff')
plt.savefig("Number_of_flights_per_continent.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python

##4 Nationality distribution of passengers

nationality_counts.plot(kind="bar", figsize=(10,5), title="Top 10 Passenger Nationalities")
plt.xlabel("Nationality")
plt.ylabel("Count")
# plt.savefig('Nationality distribution of passengers.tiff')
plt.savefig("Nationality_distribution_of_passengers.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
##5 Flight status distribution (e.g., On Time, Delayed, Cancelled)

status_counts.plot(kind="bar", color="steelblue", figsize=(8,4), title="Flight Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")
# plt.legend(loc="upper right", bbox_to_anchor=(1.15, 1))
plt.legend("")
# plt.savefig('Flight status distribution.tiff')
plt.savefig("Flight_status_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
##6 The Most frequently used airports

airport_usage.plot(kind="bar", figsize=(15,5), title="Top 10 Airports by Usage")
plt.xlabel("Airport")
plt.ylabel("Count")
# plt.xticks(rotation=90
# plt.savefig('The Most frequently used airports.tiff')
plt.savefig("The_Most_frequently_used_airports.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
##7 The cancelled flights per season

cancelled_season.plot(kind="bar", color="steelblue", figsize=(8,4))
plt.title("Cancelled Flights by Season")
plt.xlabel("Season")
plt.ylabel("Number of Cancelled Flights")
plt.grid(axis="y", linestyle="--", alpha=0.6)
# plt.savefig('The cancelled flights per season.tiff')
plt.savefig("The_cancelled_flights_per_season.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
##8 Top 10 busiest Flight Routes

# Create route labels like i.e "LAGOS → ABUJA"
top_routes["Route"] = top_routes["Airport_Name"] + " → " + top_routes["Arrival_Airport"]

# Plot
plt.figure(figsize=(12,6))
plt.barh(top_routes["Route"], top_routes["Flights"], color="cornflowerblue", edgecolor="black")
plt.xlabel("Number of Flights")
plt.ylabel("Route")
plt.title("Top 10 Busiest Flight Routes")
plt.gca().invert_yaxis()  # so the top route appears at the top
plt.grid(axis="x", linestyle="--", alpha=0.6)
# plt.savefig('Top 10 busiest Flight Routes.tiff')
plt.savefig("Top_10_busiest_Flight_Routes.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
##9 Flight Distribution by season

plt.figure(figsize=(8,5))
season_counts.plot(kind="bar", color="royalblue", edgecolor="black")
plt.title("Flight Distribution by Season")
plt.xlabel("Season")
plt.ylabel("Number of Flights")
plt.grid(axis="y", linestyle="--", alpha=0.6)
# plt.savefig('Flight Distribution by season.tiff')
plt.savefig("Flight_Distribution_by_season.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# -- Based on the exploratory data analysis conducted on the airline database, several insights have been uncovered:

# -- Gender Distribution: The dataset shows an almost equal distribution between male and female passengers. This suggests that the airline industry caters to a diverse range of travelers in terms of gender.

# -- Age Distribution: Most passengers fall within the age group of 25 to 50 years old. This indicates that the airline's services are popular among adults in their prime working and traveling years.

# -- Nationality Insights: Chinese and Indonesian nationalities dominate the dataset, implying a significant number of travelers from these countries.

# -- Country of Origin: The United States ranks as the top country of origin for travelers, indicating a high volume of inbound travelers, possibly due to its status as a major international travel hub.

# -- Travel Patterns: The prevalence of Chinese and Indonesian passengers, along with the high number of United States-originating flights, suggests a strong connection between these countries and North America.

# -- Airport Distribution: The dataset contains information on a vast number of airports, with the United States having the most airports (22104). Additionally, it's noteworthy that 40 airports share the same name but are located in different countries, highlighting the global reach of these airports.

# -- Monthly Flight Distribution: August emerges as the busiest month for flights. However, the data suggests that flight operations remain relatively consistent across the other months.

# -- Flight Status: The dataset reveals that the quantities of flight statuses (cancelled, on time, and delayed) are roughly equal. This indicates that the airline industry faces a similar distribution of these statuses, and a substantial portion of flights may experience delays or cancellations.

# -- In summary, the exploratory data analysis has shed light on various aspects of the airline industry, including passenger demographics, travel patterns, airport distribution, and flight statuses. These insights can inform strategic decisions and improvements within the airline industry to enhance customer experiences and operational efficiency.
```
