```python
# Reading the Exported Songs parquet file

parquet_filename = 'https://zenodo.org/records/8247607/files/elexon-load-20150101-20230810.parquet?download=1'
parquet_file_df = pd.read_parquet(parquet_filename, engine="fastparquet")

# Investigating the structure of the imported dataframe
parquet_file_df.info()
```

```python
# Extract the 'local_datetime','settlementDate', 'quantity' columns into a new dataframe called energy_df
```

```python
energy_df = parquet_file_df[['local_datetime','settlementDate','quantity']]
energy_df
```

```python
# Performing the necessary conversions

energy_df['settlementDate'] = pd.to_datetime(energy_df['settlementDate'])
energy_df['quantity'] = energy_df['quantity'].astype('float32') #

energy_df.info()
```

```python
# (b) Based on the date column, create thtre additional columns
# extract day, the week, and months for each data point (use google)
```

```python
## for half-hourly block data
energy_df['half_hour'] = energy_df["local_datetime"].dt.hour + energy_df["local_datetime"].dt.minute / 60

## For hourly block data
energy_df['hour'] = energy_df['local_datetime'].dt.hour
energy_df['day'] = energy_df['local_datetime'].dt.day
energy_df['week'] = energy_df['local_datetime'].dt.isocalendar().week  # ISO week number
energy_df['month'] = energy_df['local_datetime'].dt.month
energy_df['year'] = energy_df['local_datetime'].dt.year
# We have 'year', 'month', 'day', 'dayofweek', 'dayofyear', 'weekofyear', 'quarter'
energy_df
```

```python
# --- 1. Peak vs Off-Peak Loads ---
# Compare stress periods vs relaxed hours.
# Define peak hours (example: 7–10am and 5–9pm) and off-peak as others
def label_peak(hour):
    if (7 <= hour <= 10) or (17 <= hour <= 21):
        return "Peak"
    else:
        return "Off-Peak"

energy_df['period'] = energy_df['hour'].apply(label_peak)
energy_df.head()
```

```python
# --- 2. Load Duration Curve ---
# Shows reliability and high-load frequency.
# Rank loads from highest to lowest
energy_df_sorted = energy_df.sort_values(by='quantity', ascending=False).reset_index(drop=True)
energy_df_sorted['rank'] = energy_df_sorted.index + 1
```

```python
# --- 3. Load Factor ---
# Efficiency of system usage (closer to 1 = very efficient).
# Load Factor = Average Load / Peak Load
avg_load = energy_df['quantity'].mean()
peak_load = energy_df['quantity'].max()
load_factor = avg_load / peak_load

print(f"Average Load: {avg_load:.2f}")
print(f"Peak Load: {peak_load:.2f}")
print(f"Load Factor: {load_factor:.2%}")

```

```python
# --- 1. Peak vs Off-Peak Loads ---
fig, ax = plt.subplots(figsize=(8, 6))

# plt.ylim(10_000, 60_000) ## speciying lower and upper limits
# plt.xticks(rotation=90)
sns.boxplot(data=energy_df, x='period', y='quantity',widths=0.3)
ax.set_title("Peak vs Off-Peak Load Distribution")
plt.savefig("Peak_vs_Off-Peak_Loads.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# --- 2. Load Duration Curve ---
plt.figure(figsize=(10,6))
plt.plot(energy_df_sorted['rank'], energy_df_sorted['quantity'])
plt.title("Load Duration Curve")
plt.xlabel("Ranked Hours (Highest to Lowest)")
plt.ylabel("Quantity")
plt.savefig("Load_Duration_Curve.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# --- 3. Load Factor ---
# Optional: Show daily load factors in a boxplot
daily_stats = energy_df.groupby(energy_df['local_datetime'].dt.date)['quantity'].agg(['mean','max'])
daily_stats['load_factor'] = daily_stats['mean'] / daily_stats['max']

plt.figure(figsize=(6,5))
sns.boxplot(y=daily_stats['load_factor'], width=0.3)
plt.title("Distribution of Daily Load Factors", fontsize=14)
plt.ylabel("Load Factor")
plt.xlabel("")
plt.savefig("Load_Factor.png", dpi=300, bbox_inches="tight")
plt.show()
```
